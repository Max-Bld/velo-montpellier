```python
#%% Getting data from Montpellier's bike stations
import requests
import pandas as pd

api_url = "https://portail-api-data.montpellier3m.fr/bikestation?limit=1000"

response = requests.get(api_url)

response.status_code

response_dic = response.json()
```

```python
#%% Creating a map

import folium

m = folium.Map(location=(43.6081480, 3.878778), zoom_start=13)
```

```python
# Creating markers

stations = {
    "Comédie":[43.608148, 3.878778],
    "Rue Jules Ferry - Gare Saint-Roch":[43.605366, 3.881346],
    "Hôtel de Ville":[43.599088, 3.894866]
}

for station, coords in stations.items():
    folium.Marker(
        location=coords,
        tooltip=station,
        popup=f"<b>Station-vélo {station}</b><br>{coords}</br>",
        icon=folium.Icon(color='orange')).add_to(m)
```
